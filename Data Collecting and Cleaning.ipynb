```python
import json
import requests
import pandas as pd
import numpy as np
```

### Accessing the Dataset

```python
#Accessing the data from the URL

url = requests.get("https://data.cityofnewyork.us/api/views/7yc5-fec2/rows.json?accessType=DOWNLOAD")
text = url.text
print(type(text))
```

```python
#Reading the data as a JSON

data = json.loads(text)
print(type(data))
```

```python
#Preview if the data we're working with
data
```

```python
#Viewing the contents of the JSON

data.keys()
```

```python
data['meta']['view'].keys()
```

```python
print(data['meta']['view']['columns'][0])
```

### Creating a Dataframe from the Dataset

```python
#Finding the list of column headers in the dataset

for i in range(len(data['meta']['view']['columns'])):
    print(data['meta']['view']['columns'][i]['name'])
```

```python
#Number of columns in the dataset

len(data['meta']['view']['columns'])
```

```python
#Assigning the columns to a list

list1 = []
for i in range(len(data['meta']['view']['columns'])):
    list1.append(data['meta']['view']['columns'][i]['name'])
    
print(list1)
```

```python
#Finding the list of rows in the dataset

for i in range(5):
    print(data['data'][i])
```

```python
#Number of rows in the dataset

len(data['data'])
```

```python
#Assigning the rows to a list

list2 = []
for i in range(len(data['data'])):
    list2.append(data['data'][i])
    
```

```python
#Creating a dataframe from the data

df = pd.DataFrame(list2, columns = list1)
```

```python
#Viewing the dataframe

df.head()
```

```python
#Removing metadata from the dataframe

df.drop(df.columns[[0,1,2,3,4,5,6,7]],axis = 1, inplace = True)
```

### Preview of the Final dataframe obtained from the JSON

```python
#First 5 rows

df.head()
```

### Cleaning the dataframe

```python
#Listing the columns
print(df.columns.tolist())
```

```python
#Removing unwanted columns
df.drop(df.columns[[3,4,15,17,19,21,23,25,27,29,31,33,35,37,39,41,43,45,47,48,50,52,54,56,57,58,59,61,63,65,67,68,69]],axis = 1, inplace = True)
```

```python
df.head()
```

```python
#Converting junk values to NaN
df.replace("No Data", np.nan, inplace = True)
df.head()
```

```python
df.replace("s", np.nan, inplace = True)
df.head()
```

```python
print(df.columns.tolist())
```

```python
#Cleaning up the column headers
df.columns = df.columns.str.replace('[#]', '')
print(df.columns.tolist())
```

```python
df.head()
```

### Converting the dataframe from Wide to Long

```python
#Converting the dataframe from Wide to Long using melt
df_long = pd.melt(frame=df, 
          id_vars=["DBN", "School Name", "Category"], 
          var_name="Type", 
          value_name="Count")
df_long.head(10)
```

```python
#Removing NaN/Empty Values
df_cleaned = df_long[df_long['Count'].notna()]
```
